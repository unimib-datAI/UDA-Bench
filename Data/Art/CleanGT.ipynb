```python
# 检查 genre 属性值是否在值域[Portrait, Landscape, Still Life, Religious, Mythological, History, genre, Nude, Animal, Illustration, Allegorical, Abstract, Fantasy, Sketch, Figurative, Conceptual, Surrealist, Symbolic, Political, Social Realism, Pop art, Narrative, Environmental, Sculpture, Architecture, Music, Film, Digital art, Literature, Other]里，空值（包括N/A和np.nan）也视为合法
# 注意，一个cell里可能有多个值，值与值之间通过"||"隔开
# 将不在值域里的值都转换为Other

genre_accepted = [
    'Portrait', 'Landscape', 'Still Life', 'Religious', 'Mythological',
    'History', 'genre', 'Nude', 'Animal', 'Illustration', 'Allegorical',
    'Abstract', 'Fantasy', 'Sketch', 'Figurative', 'Conceptual', 'Surrealist',
    'Symbolic', 'Political', 'Social Realism', 'Pop art', 'Narrative', 'Environmental',
    'Sculpture', 'Architecture', 'Music', 'Film', 'Digital art', 'Literature', 'Other'
]

import numpy as np

# 如果 Genre 为 'N/A'，则将其替换为缺失值(np.nan)
df['Genre'] = df['Genre'].replace('N/A', np.nan)

def clean_genre_cell(cell):
    if pd.isnull(cell):
        return cell
    vals = [v.strip() for v in str(cell).split("||")]
    cleaned = [(v if v in genre_accepted else "Other") for v in vals]
    # 合并多个"Other"为一个，仅保留一个
    # 例如: ["Other", "Other", "Portrait", "Other"] 变为 ["Other", "Portrait"]
    # 处理顺序不变，仅去重相邻Other
    result = []
    seen_other = False
    for v in cleaned:
        if v == "Other":
            if not seen_other:
                result.append("Other")
                seen_other = True
        else:
            result.append(v)
    return " || ".join(result)

df['Genre'] = df['Genre'].apply(clean_genre_cell)

# 可以选择性打印被转为Other的情况
invalid_genres = []
for idx, row in df.loc[df['Genre'].notnull()].iterrows():
    vals = [v.strip() for v in str(row['Genre']).split("||")]
    for v in vals:
        if v == "Other":
            invalid_genres.append({"id": row['id'], "Name": row['Name'], "Genre": row['Genre']})
            break

if invalid_genres:
    import pandas as pd
    print("存在不在值域内的genre，已经被替换为 Other。对应的行如下：")
    print(pd.DataFrame(invalid_genres)[['id','Name','Genre']])
else:
    print("所有 genre 属性值均合法（空值也算合法）。")


```

```python
# 检查Style属性，将子类别映射为主类别，并合并重复主类别，多个主类别用"||"连接

# 1. 定义主类别和子类别的映射字典
style_map = {
    "Abstraction": [
        "Abstract Art", "Abstract Expressionism", "Action painting", "Color Field Painting", "Concretism", "Constructivism",
        "Hard Edge Painting", "Lyrical Abstraction", "Minimalism", "Neo-Concretism", "Neo-Minimalism", "Neo-Geo", "Op Art",
        "Orphism", "Post-Minimalism", "Post-Painterly Abstraction", "Spatialism", "Suprematism", "Tachisme", "Transautomatism", "Art Informel"
    ],
    "Cubism": [
        "Cubism", "Analytical Cubism", "Synthetic Cubism", "Purism"
    ],
    "Realism": [
        "Academicism", "American Realism", "Analytical Realism", "Contemporary Realism", "Naturalism", "Realism", "Regionalism",
        "Social Realism", "Socialist Realism", "Precisionism", "Luminism", "Tonalism", "Hyper-Realism", "Photorealism",
        "Fantastic Realism", "Magic Realism", "Fantasy Art", "Figurative Expressionism", "Neo-Figurative Art"
    ],
    "Impressionism": [
        "Impressionism", "Neo-Impressionism", "Post-Impressionism", "Divisionism", "Pointillism"
    ],
    "Modernism": [
        "Modernism", "Modernismo", "Neoclassicism", "Rococo", "Romanticism", "Symbolism", "Expressionism", "Fauvism",
        "Futurism", "Metaphysical art", "Surrealism", "Dada", "Lettrism", "Neo-Dada", "Neo-Expressionism", "Neo-Romanticism",
        "Neo-baroque", "New Realism", "Nouveau Réalisme", "New European Painting", "Japonism", "Orientalism"
    ],
    "Design": [
        "Art Deco", "Art Nouveau (Modern)", "Cloisonnism", "P&D (Pattern and Decoration)", "Fiber art", "Kitsch"
    ],
    "Conceptual": [
        "Conceptual Art", "Performance Art", "Feminist Art", "Postcolonial art", "Contemporary"
    ],
    "Environmental": [
        "Environmental (Land) Art", "Light and Space", "Sky Art"
    ],
    "Popular": [
        "Pop Art", "Neo-Pop Art", "Street art", "Lowbrow Art", "Junk Art"
    ],
    "Photography": [
        "Documentary photography", "Street Photography", "Pictorialism"
    ],
    "Digital": [
        "Digital Art", "Cyber Art"
    ],
    "Outsider": [
        "Art Brut", "Outsider art", "Naïve Art (Primitivism)", "Native Art"
    ],
    "Asian": [
        "Ink and wash painting", "New Ink Painting", "Nanga (Bunjinga)", "Nihonga", "Ukiyo-e", "Shin-hanga", "Sōsaku hanga", "Zen"
    ],
    "Misc": [
        "Cartographic Art", "Kinetic Art", "Muralism", "Safavid Period", "Surrealism, Muralism", "Indian Space painting"
    ]
}

# 2. 反向构造子类别到主类别的查找表
sub_to_main = {}
for main, subs in style_map.items():
    for sub in subs:
        sub_to_main[sub.strip()] = main

# 3. 编写样式规范化函数
def style_to_main(style_str):
    """将一行的风格字符串（可能由多种风格用||分隔）映射为主类别，并合并重复"""
    if pd.isna(style_str):
        return style_str
    items = [s.strip() for s in str(style_str).split("||") if s.strip()]
    main_set = set()
    for item in items:
        if item in sub_to_main:
            main_set.add(sub_to_main[item])
        else:
            # 若无法识别，保留原词
            main_set.add(item)
    return " || ".join(sorted(main_set))

# 4. 应用于DataFrame
df['Style'] = df['Style'].apply(style_to_main)

# 5. 检查是否有未覆盖的风格子类别
all_original_styles = set(s.strip() for row in df['Style'] if pd.notna(row) for s in row.split("||"))
known_main_categories = set(style_map.keys())
unrecognized_styles = all_original_styles - known_main_categories

if unrecognized_styles:
    print("以下风格未被归入任何主类别（可能为新类型/拼写误差/需进一步确认）：")
    print(unrecognized_styles)
else:
    print("所有风格均已映射至主类别。")
```

```python
# 检查Style属性是否属于值域["Abstraction","Cubism","Realism","Impressionism","Modernism","Design","Conceptual","Environmental","Popular","Photography","Digital","Outsider","Asian","Misc"]
# 空值（包括N/A和np.nan）需要转换为"Other"
valid_main_styles = {"Abstraction","Cubism","Realism","Impressionism","Modernism","Design","Conceptual","Environmental","Popular","Photography","Digital","Outsider","Asian","Misc"}

def check_and_fix_style(style_str):
    # 空值或N/A/np.nan一律转为"Other"
    if pd.isna(style_str) or str(style_str).strip().upper() in {"N/A", "NA", ""}:
        return "Other"
    # style_str 可能为 "A || B" 多类别
    styles = [s.strip() for s in str(style_str).split("||") if s.strip()]
    # 若所有子类别都在有效集合，则保留去重且原顺序组合，否则设为"Other"
    if all(s in valid_main_styles for s in styles):
        seen = set()
        dedup_styles = []
        for s in styles:
            if s not in seen:
                dedup_styles.append(s)
                seen.add(s)
        return " || ".join(dedup_styles)
    else:
        return "Other"

df['Style'] = df['Style'].apply(check_and_fix_style)
```

```python
# 检查Image_Genre属性，将子类别映射为主类别，并合并重复主类别，多个主类别用"||"连接

# 1. 定义主类别和子类别的映射字典
image_genre_map = {
    "Abstraction": ["Abstract", "Abstract Art", "Abstract Sculpture", "Non-Representational", "Minimalism"],
    "Geometry": ["Geometric", "Pattern"],
    "Conceptual": ["Conceptual"],
    "Figuration": ["Figurative", "Figure Painting", "Nude"],
    "Portraiture": ["Portrait", "Group Portrait"],
    "Landscape": ["Landscape", "Seascape", "Cityscape"],
    "Urban": ["Urban", "Urban Landscape", "Urban Scene"],
    "Street": ["Street Photography", "Street Scene", "Graffiti"],
    "Interior": ["Interior", "Interior Scene"],
    "Architecture": ["Architecture", "Architectural"],
    "Narrative": ["Narrative", "Genre", "Genre Art", "Genre Painting", "Genre Scene", "Everyday Scene", "Crowd Scene"],
    "Imagination": ["Fantasy", "Fairy Tale", "Sci-Fi", "Mythological"],
    "Religion": ["Religious"],
    "History": ["Historical", "Historical Scene"],
    "Politics": ["Political", "Political Satire"],
    "Animals": ["Animal", "Wildlife"],
    "Action": ["Action", "Sports", "Study", "Study / Motion Analysis"],
    "StillLife": ["Still Life"],
    "Medium": ["Sculpture", "Assemblage", "Mixed Media", "Installation", "Installation Art", "Mural", "Triptych", "Public Art"],
    "Graphic": ["Advertisement", "Poster", "Poster Art", "Comic", "Map Art"],
    "Industry": ["Industrial"],
    "Text": ["Text Art", "Text-Based", "Text-Based Art", "Calligraphy"]
}

# 2. 建立反向查找表：{子类别: 主类别}
sub_to_main_genre = {}
for main_cat, subs in image_genre_map.items():
    for sub in subs:
        sub_to_main_genre[sub] = main_cat

# 3. 定义转换函数
def map_image_genre_cell(cell):
    if pd.isna(cell) or str(cell).strip().upper() in {'N/A', 'NA', ''}:
        return "Other"
    genres = [g.strip() for g in str(cell).split("||") if g.strip()]
    mapped_main = []
    seen = set()
    for g in genres:
        main_cat = sub_to_main_genre.get(g)
        if main_cat and main_cat not in seen:
            mapped_main.append(main_cat)
            seen.add(main_cat)
    if mapped_main:
        return " || ".join(mapped_main)
    else:
        return "Other"

# 4. 应用到DataFrame
df['Image_Genre'] = df['Image_Genre'].apply(map_image_genre_cell)

# 可选: 打印未识别的原始子类别
all_original_image_genre = set(s.strip() for row in df['Image_Genre'] if pd.notna(row) for s in str(row).split("||"))
known_subs = set(sub_to_main_genre.keys())
not_recognized = all_original_image_genre - set(image_genre_map.keys())
if not_recognized:
    print("Image_Genre中未被识别或映射的主类别：", not_recognized)
```

```python
# 检查Image_Genre属性是否属于指定值域，不在值域及空值、N/A、NA、""均记为"Other"，"Other"保持为"Other"
valid_genres = [
    "Abstraction","Geometry","Conceptual","Figuration","Portraiture","Landscape",
    "Urban","Street","Interior","Architecture","Narrative","Imagination",
    "Religion","History","Politics","Animals","Action","StillLife",
    "Medium","Graphic","Industry","Text","Other"
]

def clean_image_genre_cell(cell):
    if pd.isna(cell) or str(cell).strip().upper() in {'N/A', 'NA', ''}:
        return "Other"
    genres = [g.strip() for g in str(cell).split("||") if g.strip()]
    cleaned = []
    for g in genres:
        if g == "Other":
            cleaned.append("Other")
        elif g in valid_genres:
            cleaned.append(g)
    # 保留唯一且顺序，不重复添加"Other"
    result = []
    seen = set()
    for g in cleaned:
        if g not in seen:
            result.append(g)
            seen.add(g)
    if not result:
        return "Other"
    return " || ".join(result)

df['Image_Genre'] = df['Image_Genre'].apply(clean_image_genre_cell)
```

```python
# 检查Color属性，如果Color为"Other"，则保持为"Other"，否则将Color映射为指定值域，多个值用"||"连接
# 注意一个cell里可能为 a||b，如果a的字符串里具有main_color，那就映射为其中的main_color，如果字符串里有多个main_color，那就以"||"分割"
main_color = ["Grey", "Gray", "Black", "White", "Yellow", "Red", "Pink", "Brown", "Orange", "Green", "Blue", "Purple", "Gold", "Silver", "Other"]

def extract_all_main_colors(s):
    result = []
    for mc in main_color:
        # 只精确包含字符串(注意防止"Red"被"Redwood"等包含)，这里简单用空格和逗号等分割，或直接in
        # 若main_color为"Other"，需全等，其他采用分词或者子串,此处直接采用分词分割或in
        if mc == "Other":
            if s.strip() == "Other":
                result.append("Other")
        else:
            # 检查独立单词模式或直接包含
            if mc in s and mc not in result:
                result.append(mc)
    return result

def clean_color_cell(cell):
    # 处理空值和特殊缺失值
    if pd.isna(cell) or str(cell).strip().upper() in {'N/A', 'NA', ''}:
        return "Other"
    # 分割颜色
    colors = [c.strip() for c in str(cell).split("||") if c.strip()]
    cleaned = []
    for c in colors:
        if c == "Other":
            if "Other" not in cleaned:
                cleaned.append("Other")
        else:
            matched_colors = extract_all_main_colors(c)
            for mc in matched_colors:
                if mc not in cleaned:
                    cleaned.append(mc)
    if not cleaned:
        return "Other"
    return " || ".join(cleaned)

# 检查是否有不在main_color中的原始值，并收集（只对未能匹配main_color的片段警告）
not_recognized_color = set()
for color_cell in df['Color'].dropna():
    values = [v.strip() for v in str(color_cell).split("||") if v.strip()]
    for v in values:
        if v == "Other":
            continue
        found = False
        for mc in main_color:
            if mc != "Other" and mc in v:
                found = True
                break
            if mc == "Other" and v.strip() == "Other":
                found = True
                break
        if not found and v.upper() not in {"N/A", "NA", ""}:
            not_recognized_color.add(v)
if not_recognized_color:
    print("Color中未被识别或未能映射的原始片段：", not_recognized_color)

df['Color'] = df['Color'].apply(clean_color_cell)
```

```python
# 检查Color属性是否属于指定值域，不在值域及空值、N/A、NA、""均记为"Other"，"Other"保持为唯一"Other"
allowed_colors = set(main_color)  # main_color应为所有允许的颜色类别

def validate_color_cell(cell):
    if pd.isna(cell) or str(cell).strip().upper() in {'N/A', 'NA', ''}:
        return "Other"
    values = [v.strip() for v in str(cell).split("||") if v.strip()]
    validated = []
    other_added = False  # 控制Other只加一次
    for v in values:
        if v == "Other":
            if not other_added:
                validated.append("Other")
                other_added = True
        elif v in allowed_colors:
            if v not in validated:
                validated.append(v)
        else:
            if not other_added:
                validated.append("Other")
                other_added = True
    if not validated:
        return "Other"
    return " || ".join(validated)

df['Color'] = df['Color'].apply(validate_color_cell)
```
